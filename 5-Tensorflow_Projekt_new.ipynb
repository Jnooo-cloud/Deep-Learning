## Daten Vorbereitung

Wenn wir Neuronale Netze und Deep Learning basierte Systeme nutzen, dann empfiehlt es sich üblicherweise die Daten zu Standardisieren. Dieser Schritt ist für unseren konkreten Datensatz nicht nötig, wir möchten aber doch einmal gemeinsam durchgehen.

### Standard Scaling

**Importiere `StandardScaler` aus `sklearn.preprocessing`.**

```python
from sklearn.preprocessing import StandardScaler
```

**Erstelle ein StandardScaler() Objekt namens "scaler".**

```python
scaler = StandardScaler()
```

**Fitte scaler auf die Features.**

```python
scaler.fit(data.drop('Class',axis=1))
```

**Nutze die `transform()` Methode, um die Features in eine skalierte Version umzuwandeln.**

```python
scaled_features = scaler.fit_transform(data.drop('Class',axis=1))
```

**Konvertiere die skalierten Features in einen DataFrame und sieh dir dessen Head an.**

```python
df_feat = pd.DataFrame(scaled_features,columns=data.columns[:-1])
df_feat.head()
```

## Train Test Split

**Erstelle zwei Objekte X und y, die die skalierten Feature-Werte und Labels beinhalten.**

```python
X = df_feat
```

```python
y = data['Class']
```

**Nutze die `to_numpy()` Methode auf X und y und überschreibe sie mit dem neuen Ergebnis.**

*Wir müssen dies tun, da Tensorflow mit Numpy Arrays arbeitet anstatt mit Pandas Series.**

```python
X = X.to_numpy()
y = y.to_numpy()
```

**Nutze SciKit Learn, um Trainings- und Testsets zu erzeugen.**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Keras Sequential Model mit TensorFlow 2.x

**Importiere die notwendigen Komponenten aus Keras. Hinweis: Das alte `tf.contrib.learn` API existiert nicht mehr in TensorFlow 2.x. Wir nutzen stattdessen die moderne Keras Sequential API, die die aktuelle Best Practice ist.**

```python
from tensorflow import keras
from tensorflow.keras import layers
```

**Erstelle ein Objekt namens "model", das ein Sequential Model ist. Füge Dense Layer mit 10, 20 und 10 Neuronen hinzu, mit 'relu' Aktivierung und einem Output Layer mit 1 Neuron und 'sigmoid' Aktivierung.**

```python
model = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(4,)),
    layers.Dense(20, activation='relu'),
    layers.Dense(10, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

**Kompiliere das Modell mit 'adam' Optimizer, 'binary_crossentropy' Loss und 'accuracy' Metrik.**

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

**Jetzt fitte das Modell auf das Trainingsset. Nutze `epochs=100` und `batch_size=20`.**

```python
model.fit(x_train, y_train, epochs=100, batch_size=20, verbose=0)
```

## Modell Auswertung

**Nutze die `predict` Methode, um Vorhersagen für x_test zu erstellen.**

```python
note_predictions = (model.predict(x_test) > 0.5).astype(int).flatten()
```

**Jetzt erstelle Classification Report und Confusion Matrix.**

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,note_predictions))
```

```python
print(classification_report(y_test,note_predictions))
```

## Optionaler Vergleich

Dir sollten extrem akkurate Ergebnisse beim DNN Modell auffallen. Vergleichen wir das mit einem Random Forest Classifier für eine Einordnung der Ergebnisse.

**Nutze SciKit Learn, um einen Random Forest Classifier zu erstellen. Vergleiche anschließend Classification Report und Confusion Matrix.**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc_preds = rfc.predict(x_test)
```

```python
print(classification_report(y_test,rfc_preds))
```

```python
print(confusion_matrix(y_test,rfc_preds))
```
