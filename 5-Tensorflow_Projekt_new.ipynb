# TensorFlow Projekt

Banknote-Daten

```python
import pandas as pd
import tensorflow as tf
print(tf.__version__)
```

```python
data = pd.read_csv('bank_note_data.csv')
```

head()

```python
data.head()
```

## EDA

Imports

```python
import seaborn as sns
%matplotlib inline
```

Countplot: `Class`

```python
sns.countplot(x='Class',data=data)
```

Pairplot mit `Class`

```python
sns.pairplot(data,hue='Class')
```

## Daten vorbereiten

```python
from sklearn.preprocessing import StandardScaler
```

`StandardScaler()` erstellen

```python
scaler = StandardScaler()
```

Scaler fitten

```python
scaler.fit(data.drop('Class',axis=1))
```

Features transformieren

```python
scaled_features = scaler.fit_transform(data.drop('Class',axis=1))
```

Skalierte Daten prüfen

```python
df_feat = pd.DataFrame(scaled_features,columns=data.columns[:-1])
df_feat.head()
```

## Train/Test Split

`X` und `y` erstellen

```python
X = df_feat
```

```python
y = data['Class']
```

`to_numpy()` auf `X` und `y`

```python
X = X.to_numpy()
y = y.to_numpy()
```

Train/Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Keras-Modell

```python
from tensorflow import keras
from tensorflow.keras import layers
```

Sequential-Modell erstellen

```python
model = keras.Sequential([
    layers.Dense(10, activation='relu', input_shape=(4,)),
    layers.Dense(20, activation='relu'),
    layers.Dense(10, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

Modell kompilieren

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

Modell fitten

```python
model.fit(x_train, y_train, epochs=100, batch_size=20, verbose=0)
```

## Auswertung

Vorhersagen

```python
note_predictions = (model.predict(x_test) > 0.5).astype(int).flatten()
```

Classification Report und Confusion Matrix

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,note_predictions))
```

```python
print(classification_report(y_test,note_predictions))
```

## Vergleich

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc_preds = rfc.predict(x_test)
```

```python
print(classification_report(y_test,rfc_preds))
```

```python
print(confusion_matrix(y_test,rfc_preds))
```
